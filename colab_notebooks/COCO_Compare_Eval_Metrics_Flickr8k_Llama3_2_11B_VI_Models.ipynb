```python
# Create a dictionary mapping image name to a list of captions
image_captions = {}
for annotation in captions_data['annotations']:
    image_id = annotation['image_id']
    caption = annotation['caption']
    if image_id not in image_captions:
        image_captions[image_id] = []
    image_captions[image_id].append(caption)

# List image files
images_dir = '/content/val2017/'
image_files = [f for f in os.listdir(images_dir) if f.endswith('.jpg')]

print(f"Loaded {len(captions_data['annotations'])} captions for {len(image_captions)} images.")
print(f"Found {len(image_files)} image files in {images_dir}.")
```

```python
imagename_captions = dict()
for images in captions_data["images"]:
  imagename = images['file_name']
  imageid = images['id']
  #imagename_captions[imagename] = imageid
  #print(imagename, imageid)
  i = 0
  for annotation in captions_data['annotations']:
    if annotation['image_id'] == imageid:
      caption = annotation['caption']
      if imagename+"#"+str(i) not in imagename_captions.keys():
        imagename_captions[imagename+"#"+str(i)] = caption
      i += 1
```

```python
imagename_captions
```

```python
IMAGES_PATH = "/content/val2017/"
text_data = []
caption_mappings = {}
```

```python
def load_captions_data():
    for imagename in imagename_captions.keys():
        caption = imagename_captions[imagename]
        img_name = imagename.split("#")[0]
        img_name = os.path.join(IMAGES_PATH, img_name.strip())

        if img_name.endswith("jpg"):
          caption = "<start> " + caption.strip() + " <end>"
          text_data.append(caption)
          if img_name in caption_mappings:
            caption_mappings[img_name].append(caption)
          else:
            caption_mappings[img_name] = [caption]
    return caption_mappings, text_data
```

```python
captionsmappings, text_data = load_captions_data()
```

```python
def validation_data(captionsmappings):
    # 1. Get the list of all image names
    all_images = list(captionsmappings.keys())
    validation_data = {
        img_name: captionsmappings[img_name] for img_name in all_images
    }
    # 4. Return the splits
    return validation_data
```

```python
validation_data = validation_data(captionsmappings)
```

```python
validation_data
```

```python
instruction = "Caption this image in less than 12 words"

def convert_to_conversation(sample):
    conversation = [
        { "role": "user",
          "content" : [
            {"type" : "text",  "text"  : instruction},
            {"type" : "image", "image" : sample["image"]} ]
        },
        { "role" : "assistant",
          "content" : [
            {"type" : "text",  "text"  : sample["caption"]} ]
        },
    ]
    return { "messages" : conversation }
pass
```

```python
converted_dataset = []
for image_path, captions in validation_data.items():
  for caption in captions:
    # Create a dictionary in the format expected by convert_to_conversation
    sample = {"image": image_path, "caption": caption}
    converted_dataset.append(convert_to_conversation(sample))
```

```python
def compute_scores(model_name, predicted_captions, text_data, test_size):
  results = {}
  #results[model_name] = model_name
  results["wer_score"] = wer.compute(predictions=predicted_captions, references=text_data[:test_size])
  results["bleu_score"] = bleu.compute(predictions=predicted_captions, references=text_data[:test_size])
  results["rouge_score"] = rouge.compute(predictions=predicted_captions, references=text_data[:test_size])
  results["meteor_score"] = meteor.compute(predictions=predicted_captions, references=text_data[:test_size])
  results["perplexity_score"] = perplexity.compute(predictions=predicted_captions, references=text_data[:test_size], model_id=model="meta-llama/Llama-3.2-11B-Vision-Instruct")
  return results
```

```python
test_size = 100

def get_predicted_captions(filename,model,tokenizer):
  predicted_captions = []
  images = list(validation_data.keys())
  instruction = "Caption this image in less than 12 words"
  messages = [
    {"role": "user", "content": [
        {"type": "image"},
        {"type": "text", "text": instruction}
    ]}
  ]
  input_text = tokenizer.apply_chat_template(messages, add_generation_prompt = True)
  with open(filename, "w") as f:
    i = 0
    for image in images:
      im = Image.open(image)
      inputs = tokenizer(
        im, # Pass the image path as a list
        input_text,
        add_special_tokens = False,
        return_tensors = "pt").to("cuda")
      text_streamer = TextStreamer(tokenizer, skip_prompt = True)
      _ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 128,
                   use_cache = True, temperature = 1.5, min_p = 0.1)
      predicted_caption = " ".join(tokenizer.decode(_[0], skip_special_tokens=True)) # Decode the generated tokens
      #predicted_captions.append(predicted_caption)
      f.write(predicted_caption)
      f.write("\n")
      i += 1
      if i == test_size:
        break
pass
```

```python
# Thanks ChatGPT
import matplotlib.pyplot as plt
import numpy as np

# Metric names
labels = ['WER', 'BLEU', 'ROUGE-1', 'ROUGE-2', 'ROUGE-L', 'ROUGE-Lsum', 'METEOR']

# Sample Raw scores
r1_scores = [
    1.9886731391585761,
    0.0,
    0.022969240085673442,
    0.0,
    0.023217138976832183,
    0.023243251688960624,
    0.04750829752391568
]

r2_scores = [
    1.986731391585761,
    0.0,
    0.322969240085673442,
    0.01,
    0.065217138976832183,
    0.029843251688960624,
    0.12750829752391568
]

# Invert WER because lower is better
r1_scores[0] = 1 / r1_scores[0]
r2_scores[0] = 1 / r2_scores[0]

# Normalize scores (0 to 1 range)
max_vals = np.maximum(r1_scores, r2_scores)
r1_norm = np.array(r1_scores) / max_vals
r2_norm = np.array(r2_scores) / max_vals

# Radar setup
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
r1_plot = np.concatenate((r1_norm, [r1_norm[0]]))
r2_plot = np.concatenate((r2_norm, [r2_norm[0]]))
angles += angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.plot(angles, r1_plot, label='r1', linewidth=2)
ax.fill(angles, r1_plot, alpha=0.25)

ax.plot(angles, r2_plot, label='r2', linewidth=2)
ax.fill(angles, r2_plot, alpha=0.25)

ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_title('Radar Chart: Normalized Comparison of r1 vs r2', size=14)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()
```
